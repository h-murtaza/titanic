```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

TD = pd.read_csv("titanic-passengers.csv", ';')

TD.head()
```

```python
TD.info()
```

```python
TD['Age'].mean()
```

```python
TD['Age'] = TD['Age'].fillna(TD['Age'].mean())
```

```python
TD.drop('Cabin',axis=1, inplace=True)
```

```python
TD.info()
```

```python
TD['Embarked'] = TD['Embarked'].fillna('S')
```
